# Stock Market Time Series Analysis

### Objective
Analyze historical stock price data to understand trends, volatility, and return patterns using Python.

### Tools Used
- Python
- Pandas
- NumPy
- Matplotlib
- Seaborn

```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Dataset
df = pd.read_csv("Stock_Data.csv")
df.head()
```

```python
#Dataset Information
df.info()
df.describe()
```

```python
#Convert Date Column
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
df.set_index('Date', inplace=True)

df.head()
```

```python
#Closing Price Trend
plt.figure(figsize=(10,5))
plt.plot(df['Close'])
plt.title("Closing Price Trend")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.show()
```

### Insight
The chart shows the long-term trend of the stock price. Upward trends indicate market growth while downward trends may reflect market corrections.

```python
#Daily Return
df['Daily Return'] = df['Close'].pct_change()

df[['Close','Daily Return']].head()
```

```python
#Visualization
plt.figure(figsize=(10,5))
plt.plot(df['Daily Return'])
plt.title("Daily Returns")
plt.show()
```

### Insight
Daily returns represent the percentage change in stock price from one day to the next. Large spikes indicate higher volatility.

```python
#Moving Averages
df['MA20'] = df['Close'].rolling(window=20).mean()
df['MA50'] = df['Close'].rolling(window=50).mean()
```

```python
#Visualization
plt.figure(figsize=(10,5))
plt.plot(df['Close'], label='Close Price')
plt.plot(df['MA20'], label='20 Day MA')
plt.plot(df['MA50'], label='50 Day MA')

plt.legend()
plt.title("Moving Averages")
plt.show()
```

### Insight
Moving averages smooth short-term fluctuations and highlight long-term trends. When the short-term average crosses the long-term average, it may indicate trend changes.

```python
# Volatility Analysis
df['Volatility'] = df['Daily Return'].rolling(window=20).std()
```

```python
#Visualization
plt.figure(figsize=(10,5))
plt.plot(df['Volatility'])
plt.title("20-Day Rolling Volatility")
plt.show()
```

### Insight
Volatility measures the level of price fluctuation. Higher volatility indicates higher market uncertainty and risk.

```python
#Trading Volume Analysis
top_volume = df.sort_values('Volume', ascending=False).head(10)

top_volume[['Close','Volume']]
```

### Insights
This identifies days with unusually high trading activity.

# Key Takeaways

- The stock shows long-term growth trend.
- Daily returns highlight volatility patterns.
- Moving averages help identify potential trend reversals.
- Periods of high volatility correspond with market uncertainty.
